```python
import pandas as pd
import numpy as np
import re 
pd.set_option('display.max_columns', None)
```

```python
df_part1 = pd.read_csv("archivos_pair/world-data-2023_part1.csv",index_col=0)
df_part2 = pd.read_csv("archivos_pair/world-data-2023_part2.csv",index_col=0)
```

```python
df_merge = df_part1.merge(df_part2, how='inner' ,left_on="Country", right_on="country")
```

```python
df_merge.head(2)
```

```python
df_merge.drop("country",axis=1,inplace=True)
```

```python

nuevas_columnas = {columna: columna.lower().replace(":", "_").replace(" ", "_").replace("\n","") for columna in df_merge.columns} 


df_merge.rename(columns = nuevas_columnas, inplace = True)
```

```python
def limpiar_nombre_columna(nombre):  
    nombre = re.sub(r'\(.*?\)', '', nombre) 
    return nombre.strip()
```

```python
nuevas_columnas = {col: limpiar_nombre_columna(col) for col in df_merge.columns}
```

```python
df_merge.rename(columns=nuevas_columnas, inplace=True)
```

```python
df_merge.head(1)
```

```python
df_merge[["longitud", "latitud"]] = df_merge["coordinates"].str.split(",", expand=True).get([0, 1])
```

```python
df_merge.drop("coordinates",axis=1,inplace=True)
```

```python
df_merge.head(1)
```

```python
df_merge['longitud'] = df_merge['longitud'].str.replace("(", "").str.replace("'", "").str.strip()
df_merge['latitud'] = df_merge['latitud'].str.replace(")", "").str.replace("'", "").str.strip()
```

```python
nuevas_columnas = ['unemployment_rate', 'total_tax_rate', 'tax_revenue_', 'population__labor_force_participation_', 'out_of_pocket_health_expenditure', 'gross_tertiary_education_enrollment_', 'gross_primary_education_enrollment_', 'forested_area_', 'cpi_change_', 'agricultural_land']


df_merge[nuevas_columnas] = df_merge[nuevas_columnas].str.replace('%', '')
```

```python
df_merge['cpi_change_'] = df_merge['cpi_change_'].str.replace('%', '')
```
